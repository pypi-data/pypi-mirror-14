```python
%matplotlib inline
import scipy.sparse as sp
from scipy.io import loadmat
import numpy as np
from coclust.CoclustMod import CoclustMod
import matplotlib.pyplot as plt

def shuffled_index(m) :
  index=np.arange(np.shape(m)[0])
  np.random.shuffle(index)
  return index


# Retrieve the CSTR  document-term matrix from a matlab file
file_name = "../datasets/cstr.mat"
matlab_dict = loadmat(file_name)
X = matlab_dict['fea']
print(X.shape)

# Shuffle the matrix
X_shuff=X[shuffled_index(X),:]
X_1=X_shuff[:158]
X_2=X_shuff[:316]
X_3=X_shuff[316:]

# Create and fit a model with 4 co-clusters
model1 = CoclustMod(n_clusters=4)
model1.fit(X_1)
print("Ok")
model2 = CoclustMod(n_clusters=4)
model2.fit(X_2)
print("Ok")
model3 = CoclustMod(n_clusters=4)
model3.fit(X_3)
print("Ok")
X_1_fit= X_1[np.argsort(model1.row_labels_)]
X_1_fit = X_1_fit[:, np.argsort(model1.column_labels_)]
plt.spy(X_1_fit,precision=0.8, markersize=0.9)
plt.show()

X_2_fit= X_2[np.argsort(model2.row_labels_)]
X_2_fit = X_2_fit[:, np.argsort(model2.column_labels_)]
plt.spy(X_2_fit,precision=0.8, markersize=0.9)
plt.show()

X_3_fit= X_3[np.argsort(model3.row_labels_)]
X_3_fit = X_3_fit[:, np.argsort(model3.column_labels_)]
plt.spy(X_3_fit,precision=0.8, markersize=0.8)
plt.show()


```
