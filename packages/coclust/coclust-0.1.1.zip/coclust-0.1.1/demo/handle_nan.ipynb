```python
X=np.array([[0, 1, 0],
       [1, 0, 1],
       [0, 0, 0],
       [1, 1, 0],
       [0, 0, 0]])

X.any()   # True
print(X.any(axis=1)) # [ True  True False  True False] ; True = la ligne a une val. diff. de 0
~X.any(axis=1) # = array([False, False,  True, False,  True], dtype=bool)
np.where(~X.any(axis=1))[0]  # array([2, 4], dtype=int64) ; np.where donne les index des True

X.all(axis=1) # False, False, False, False, False] ;aucune ligne complete
```
